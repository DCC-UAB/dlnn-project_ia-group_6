```python
def audio_labeled(audio_dir, output_dir, song_dict, genre_dict):
    for folder_name in os.listdir(audio_dir):
        folder_path = os.path.join(audio_dir, folder_name)

        if not os.path.isdir(folder_path):
            continue
            

        for audio_file in glob.glob(os.path.join(folder_path, "*.mp3")):
            file_name = os.path.basename(audio_file)
            numbers, _ = file_name.split(".")
            song_type = song_dict.get(int(numbers))
            

            if song_type is None:
                continue

            type_code = genre_dict.get(song_type)
            
            if type_code is None:
                continue

            new_file_name = str(type_code) + "_" + file_name

            new_file_path = os.path.join(output_dir, new_file_name)

            os.rename(audio_file, new_file_path)
```

```python
# dictionary mapping song types to genre codes
genre_dict = {"Electronic" :0, "Experimental" :1, "Folk" :2, "Hip-Hop":3, "Instrumental":4, "International":5, "Pop": 6, "Rock": 7}
x = list(genre_dict.keys())

tracks = tracks[tracks['genre_top'].isin(x)]

genre_top_ = tracks['genre_top']
track_id_ = raw_tracks['track_id']

all_genres = genre_top_.values.T.tolist()
all_song_id = track_id_.values.T.tolist()

# dictionary mapping song numbers to their corresponding song types
song_dict = dict(zip(all_song_id, all_genres))
```

```python
import os

input_dir = "/Users/noramendoza/Desktop/DL_project/ALL_DATA/fma_small" # directory path where the audio files are located

output_dir = "/Users/noramendoza/Desktop/DL_project/ALL_DATA/label" # directory path where the labeled audio files will be saved
os.mkdir(output_dir) # create new directory

audio_labeled(input_dir, output_dir, song_dict, genre_dict)
```

```python
import json

file_path = "/Users/noramendoza/Desktop/DL_project/ALL_DATA/song_dictionary.json"

# Open the file in write mode
with open(file_path, "w") as file:
    # Write the dictionary as JSON to the file
    json.dump(song_dict, file)
```
